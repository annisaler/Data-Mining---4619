# Implementation dataset_iris.csv

```python
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
iris = pd.read_csv('/content/sample_data/dataset_iris.csv', delimiter=';', header=0)
```

```python
iris.head()
```

Mengubah kolom "Species" menjadi Unique - Interger 

```python
iris["Species"] = pd.factorize(iris.Species)[0]
```

```python
iris.head()
```

```python
print(iris)
```

```python
iris = iris.drop(labels="Id", axis=1)
```

```python
print(iris)
```

```python
iris = iris.to_numpy()
print(iris)
```

```python
dataTraining = np.concatenate((iris[0:40, :], iris[50:90]),
                              axis=0)
dataTesting = np.concatenate((iris[40:50, :], iris[90:100]),
                              axis=0)
print(dataTraining)
len(dataTraining)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
inputTraining = dataTraining[:, 0:4]
inputTesting = dataTesting[:, 0:4]
labelTraining = dataTraining[:, 4]
labelTesting = dataTesting[:, 4]
print(labelTraining)
len(labelTraining)
```

```python
model = model.fit(inputTraining, labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Sebenarnya : ", hasilPrediksi)
```

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print(prediksiBenar)
print(prediksiSalah)
print(prediksiBenar/(prediksiBenar+prediksiSalah) * 100)
```
